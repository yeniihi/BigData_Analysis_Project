```python
popular = pd.read_csv('./data/서울시/인구밀도_20221109153406.csv', encoding='utf-8')

# 칼럼 정리
popular.columns = list(popular.iloc[0, :])

# 불필요한 데이터 제거
popular = popular.iloc[1:, [1, 2, 3]]
spac = popular[popular['동별(3)'].str.contains('소계')].index
popular.drop(spac, inplace=True)
popular.sort_values('동별(2)', inplace=True)

popular
```

```python
# 법정동 row 정리
popular['동별(3)'] = popular['동별(3)'].str.replace('[^가-힣]', '')

# 인구 (명) 
popular['인구 (명)'] = popular['인구 (명)'].astype('int')

# 칼럼 정리
popular.columns = ['자치구', '법정동', '인구(명)']
```
